```python
import pandas as pd

df = pd.read_csv('customer_shopping_behavior.csv')
```

```python
df.head()  #check top 5 rows
df.describe()  #check statistics summary of numerical columns only
```

```python
df.info(10)  #check data types
```

```python
df.describe(include='all')  #check statistics summary of numerical columns and categorical columns both
```

```python
df.isnull().sum() #gives count of null values for columns
#aim is to replaced null ratings with median value but just changing it to median is not a good idea because each category has different median 
#and may vary introducing bias, hence we will use median review rating of each category and replace it with missing values according to category
```

```python
#replacing NULL with median according to category
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
#df.groupby(‘category’) groups rows by unique values in ‘Category’ column
# [review rating] slects only review rating column within each group
# transform used to change, lamda creates function x which is filled with median
```

```python
#cleaning column names to lower case and removing special characters
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
df = df.rename(columns={'purchase_amount_(usd)' : 'purchase_amount'})
```

```python
df.head()
```

```python
#create a column age_group
labels = ['Young adult', 'adult', 'middle-aged', 'senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)
#divides age into 4 equal groups based on number of data and name these group labels as labels defined above
```

```python
df[['age','age_group']].head()
```

```python
#create column purchase_frequency_days
#this is in text format like weekly, quarterly etc but its difficult to do analysis without numerical data
#Make sure upper case and lower case is same otherwise column values will show NaN
frequency_mapping = {
'Fortnightly':14,
'Weekly':7,
'Monthly':30,
'Quarterly':90,
'Bi-Weekly':14,
'Annually':365,
'Every 3 Months':90
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head()
```

```python
(df['discount_applied'] == df['promo_code_used']).all()

#Used to check if d=both columns have same value as Yes filled. Because there may be chances that discount may be given due to membership/cards 
#without any promo code. Since both these columns are same. We discard promo code used column
```

```python
df=df.drop('promo_code_used', axis = 1)
```
